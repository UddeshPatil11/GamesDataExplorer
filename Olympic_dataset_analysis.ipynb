```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sns
athlete = pd.read_csv('/content/athlete_events.csv')
region =pd.read_csv('/content/noc_regions.csv')
```

```python
athlete.head()
```

```python
region.head()
```

```python
#join the dataframe

athlete_df =athlete.merge(region, how ="left", on="NOC")
athlete_df.head()
```

```python
athlete_df.shape
```

```python
#column names consistent

athlete_df.rename(columns={'region': 'Region','notes': 'Notes'},inplace=True);
athlete_df.head()
```

```python
#Top countries partipicating
top_10_countries=athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#plot for the top 10 countries

plt.figure(figsize=(12,6))

plt.title('Overall percentage by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette ="Set2");
```

```python
#age distribution
plt.figure(figsize=(12,6))
plt.title('Overall Age distribution to athlete')
plt.xlabel('Age')
plt.ylabel('NUmber of participants')
plt.hist(athlete_df.Age,bins = np.arange(10,80,2), color='orange', edgecolor='white')
```

```python
#counting the gender value
gender_count=athlete_df.Sex.value_counts()
gender_count
```

```python
#pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_count, labels =gender_count.index ,autopct='%1.1f%%',startangle =150, shadow =True);
```

```python
athlete_df.Medal.value_counts()
```

```python
#total number of female athletes in each olympic
female_participant= athlete_df[(athlete_df.Sex=='F') & (athlete_df.Season=='Summer')][['Sex','Year']]
female_participant = female_participant.groupby('Year').count().reset_index()
female_participant.tail()
```

```python
#applying to check female_athlete in summer
womenOlympic =athlete_df[(athlete_df.Sex=='F') & (athlete_df.Season=='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenOlympic , palette='Spectral')
plt.title("Women Participation")
```

```python
part =womenOlympic.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title("Plot of Female Athletes over time")
```

```python
#Gold medal athletes
goldMedals = athlete_df[(athlete_df.Medal=="Gold")]
goldMedals.head()
```

```python
#take only the values that are differnt from Nan
goldMedals =goldMedals[np.isfinite(goldMedals['Age'])]
goldMedals
```

```python
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
#plot for sporting_event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for athletes over 60 years')
```

```python
#gold medals frome each country
goldMedals.Region.value_counts().reset_index(name ="Medal").head(5)
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name= "Medal").head(5)
g=sns.catplot(x="index",y="Medal", data =totalGoldMedals,
              height=5,kind="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels('Top 5 Country')
g.set_ylabels('number medals')
plt.title('Gold Medals as per country')
```
